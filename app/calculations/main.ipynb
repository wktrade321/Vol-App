```python
spy = yf.download('SPY', start=datetime(2003,1,1), progress=False)
qqq = yf.download('QQQ', start=datetime(2003,1,1), progress=False)
```

```python
spy['return_close_to_open'] = spy['Open']/spy['Close'].shift(1) - 1
spy['return_open_to_close'] = spy['Close']/spy['Open'] - 1
spy['return_close_to_close'] = spy['Close'].pct_change(1)
spy['return_open_to_open'] = spy['Open'].pct_change(1)
spy['return_cc_lag'] = spy['return_close_to_close'].shift(1)
spy['return_oo_lag'] = spy['return_open_to_open'].shift(1)
```

```python
qqq['return_close_to_open'] = qqq['Open']/qqq['Close'].shift(1) - 1
qqq['return_open_to_close'] = qqq['Close']/qqq['Open'] - 1
qqq['return_close_to_close'] = qqq['Close'].pct_change(1)
qqq['return_open_to_open'] = qqq['Open'].pct_change(1)
qqq['return_cc_lag'] = qqq['return_close_to_close'].shift(1)
qqq['return_oo_lag'] = qqq['return_open_to_open'].shift(1)
```
